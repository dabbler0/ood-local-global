```python
max_batch_size = 128

criterion = torch.nn.CrossEntropyLoss()

def run_test(m, chunk = chunk0):
    global max_batch_size
    valid_loss = 0
    samples = 0
    
    for length in tqdm(chunk):
        # Shuffle the pool of sentences of each length
        pool = chunk[length]

        # Divide into batches of at most batch_size
        for i in range(len(pool) // max_batch_size + 1):
            if i * max_batch_size < len(pool):
                batch = torch.LongTensor(
                    pool[i * max_batch_size : (i + 1) * max_batch_size]
                ).cuda()

                # Test on these batches
                batch_size, seq_len = batch.shape

                hidden = m.init_hidden(batch_size).cuda()

                for i in range(seq_len - 1):
                    output, hidden = m(batch[:, i], hidden)
                    valid_loss += criterion(output, batch[:, i + 1]).data * batch_size
                    samples += batch_size

    valid_loss /= samples
    print('total samples', samples)
    return valid_loss
```

```python
def seq2sen(seq, vocab = vocab):
    return ' '.join([
        vocab[c - 2][0] if c >= 2 else
        '<unk>' if c == 0 else
        '<end>'
        for c in seq
    ])
def sen2seq(sentence, vocab = vocab):
    vocab_dict = { v[0]: i + 2 for i, v in enumerate(vocab) if i < 2 ** 14 - 2 }
    tokens = [token.lower() for token in tokenizer.tokenize(sentence)]
    inds = [vocab_dict[t] if t in vocab_dict else 0 for t in tokens]
    return torch.LongTensor(inds)
```

```python
def make_rand_pool(chunk):
    rand_pool = []
    for length in tqdm(chunk):
        for seq in chunk[length]:
            trunc_len = np.random.randint(1, length)
            new_token = np.random.randint(2, 200)
            rand_pool.append(seq[:trunc_len] + [new_token])
            
    return rand_pool
```

```python
def test_hypotheses_for_model(m, subpool = subpool, bigrams = bigrams):
    outedge_preds = [ preds.outedge(m, torch.LongTensor(seq).unsqueeze(0), [0, 1], 15) for seq in tqdm(subpool) ]
    bigram_preds = [ bigram_next(seq, bigrams = bigrams) for seq in tqdm(subpool) ]
    actual_preds = [ preds.pred(m, torch.LongTensor(seq).unsqueeze(0))[0] for seq in tqdm(subpool) ]
    actual_preds = [ F.softmax(x.squeeze(), dim=0) for x in actual_preds ]
    
    def aggregate_test(a, b):
        return sum(
            torch.abs(a * outedge + b * bigram - actual).sum()
            for outedge, bigram, actual in zip(outedge_preds, bigram_preds, actual_preds)
        ) / len(outedge_preds)
    def aggregate_log_test(a, b):
        return sum(
            torch.abs(F.softmax(a * torch.log(outedge) + b * torch.log(bigram)) - actual).sum()
            for outedge, bigram, actual in zip(outedge_preds, bigram_preds, actual_preds)
        ) / len(outedge_preds)
        plt.plot([a / 100 for a in range(100)], [aggregate_test(a / 100, 1 - a / 100) for a in range(100)])
    plt.show()
    
    return outedge_preds, actual_preds
```

```python
def next_words(model, sentence, vocab = vocab):
    output, hidden = preds.pred(model, sen2seq(sentence, vocab = vocab).unsqueeze(0))
    return F.softmax(output.squeeze(), dim=0)
def next_words_seq(model, seq):
    output, hidden = preds.pred(model, seq.unsqueeze(0))
    return F.softmax(output.squeeze(), dim=0)
def topn(x, n, vocab = vocab):
    maxs, indices = torch.topk(x, n)
    return seq2sen(indices, vocab = vocab)
```

```python
def l1d(a, b, true, t):
    magnitudes = a - b
    signs = torch.sign(t * a + (1 - t) * b - true)
    return -torch.sum(magnitudes * signs)

def find_optimal_interpolation(a, b, true):
    a = a.squeeze()
    b = b.squeeze()
    true = true.squeeze()
    t = 0.5
    n = 1
    for _ in range(10):
        n += 1
        dv = l1d(a, b, true, t)
        if dv > 0:
            t += 1 / (2 ** n)
        else:
            t -= 1 / (2 ** n)
    return t, torch.abs(t * a + (1 - t) * b - true).sum(dim=1).mean()
```
